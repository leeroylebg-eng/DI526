## Exercise 1 — Small Quiz: Deep Learning Fundamentals

**Q1. What is the key difference between traditional ML and deep learning?**  
Traditional ML requires **manual feature engineering** — a human must decide which features (e.g., pixel edges, color histograms) to extract from raw data. Deep learning uses multi-layer neural networks that **automatically learn hierarchical features** directly from raw data. This allows deep learning to handle unstructured data (images, audio, text) that would be very hard to engineer features for manually.

**Q2. How do ANNs mimic the human brain?**  
The human brain consists of billions of **neurons** connected by **synapses**. An ANN mimics this with artificial neurons (nodes) connected by weighted edges. Like biological neurons that fire when a threshold is exceeded, artificial neurons apply an **activation function** to decide whether to pass a signal forward. Multiple layers of neurons allow complex, hierarchical representations — similar to how our visual cortex processes edges before shapes before objects.

**Q3. Why does deep learning perform better on large datasets?**  
Deep learning models have millions of parameters (weights) that need enough data to be tuned correctly. With small datasets, simpler models (SVM, Random Forest) generalize better because deep networks **overfit** — they memorize the training data instead of learning patterns. With large datasets (100k+ samples), deep networks have enough signal to learn rich, generalizable representations, outperforming traditional ML by a large margin.

**Q4. Challenges of deep learning and solutions?**

| Challenge | Solution |
|---|---|
| Requires huge labelled datasets | Data augmentation, transfer learning |
| High computational cost | GPUs/TPUs, cloud computing |
| Black box / low interpretability | Explainability tools (SHAP, Grad-CAM) |
| Overfitting | Dropout, L2 regularization, early stopping |
| Long training time | Pre-trained models (fine-tuning) |

**Q5. What is feature engineering, and why is it not needed in deep learning?**  
Feature engineering is the process of manually transforming raw data into meaningful inputs for a model (e.g., extracting "edge density" from an image). In deep learning, the hidden layers act as **automatic feature extractors** — early layers learn low-level features (edges), middle layers learn mid-level features (shapes), and deeper layers learn high-level features (objects). This eliminates the need for manual domain expertise.

**Q6. What role do hidden layers play?**  
Hidden layers are intermediate layers between the input and output. Each hidden layer learns a **transformation of the previous layer's representation**, enabling the model to build increasingly abstract features. Without hidden layers, the network can only learn linear relationships (like logistic regression). With multiple hidden layers, it can approximate any complex function (Universal Approximation Theorem).

**Q7. What is the function of an activation function in an ANN?**  
Without activation functions, stacking multiple linear layers would still produce a linear function — no matter how deep the network, it would be equivalent to a single linear layer. Activation functions introduce **non-linearity**, allowing the network to learn complex, non-linear decision boundaries. Common choices: ReLU for hidden layers (avoids vanishing gradients), Softmax for multi-class output (produces probabilities that sum to 1).

## Exercise 2 — Building a Simple Perceptron Decision System

```python
# Perceptron: should I go outside?
# Inputs: Temperature (°F) and Rain (0=No, 1=Yes)
# Weights: w_temp=0.6, w_rain=0.4, Bias=2
# Activation: Step function — output 1 if weighted sum > 20, else 0

def perceptron(temperature, rain, w_temp=0.6, w_rain=0.4, bias=2, threshold=20):
    weighted_sum = (temperature * w_temp) + (rain * w_rain) + bias
    output = 1 if weighted_sum > threshold else 0
    decision = "Yes, go outside! 🌤️" if output == 1 else "No, stay inside! 🏠"
    return weighted_sum, output, decision

cases = [
    (70, 0, "Case 1: Temperature=70°F, Rain=No"),
    (50, 1, "Case 2: Temperature=50°F, Rain=Yes"),
]

print("=" * 55)
print("     PERCEPTRON DECISION SYSTEM — Go Outside?")
print("=" * 55)
for temp, rain, label in cases:
    ws, out, decision = perceptron(temp, rain)
    print(f"\n{label}")
    print(f"  Weighted Sum = ({temp}×0.6) + ({rain}×0.4) + 2 = {ws}")
    print(f"  Output       = {out}  →  {decision}")

print("\n" + "=" * 55)
print("INTERPRETATION:")
print("""
Case 1 (70°F, No Rain):
  Weighted Sum = 42 + 0 + 2 = 44 > 20 → OUTPUT = 1 ✓
  The temperature is warm and there's no rain, so the
  perceptron correctly suggests going outside.

Case 2 (50°F, Rain):
  Weighted Sum = 30 + 0.4 + 2 = 32.4 > 20 → OUTPUT = 1
  Even with rain, 50°F is mild enough that the high
  temperature weight dominates. The perceptron still
  suggests going outside — though in reality rain at
  50°F might be unpleasant. This shows a limitation:
  simple perceptrons can't model complex interactions
  between features (e.g., cold + rain = definitely stay in).
""")
```

## Exercise 3 — Neural Network with TensorFlow/Keras (MNIST)

```python
# Step 1 — Import TensorFlow and Keras modules
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt

print(f"TensorFlow {tf.__version__} ✓")
```

```python
# Step 2 — Load the MNIST Dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(f"Training images: {X_train.shape}  Labels: {y_train.shape}")
print(f"Test images:     {X_test.shape}  Labels: {y_test.shape}")
print(f"Pixel range before normalization: [{X_train.min()}, {X_train.max()}]")

# Step 3 — Normalize pixel values to [0, 1]
X_train = X_train / 255.0
X_test  = X_test  / 255.0
print(f"Pixel range after  normalization: [{X_train.min():.1f}, {X_train.max():.1f}] ✓")

# Step 4 — One-Hot Encode the Labels
y_train_ohe = to_categorical(y_train, num_classes=10)
y_test_ohe  = to_categorical(y_test,  num_classes=10)
print(f"\nLabel before OHE: {y_train[0]}  →  After OHE: {y_train_ohe[0]}")
```

```python
# Step 5 — Build the Neural Network
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),          # 28x28 → 784
    layers.Dense(128, activation='relu'),           # Hidden layer: 128 neurons + ReLU
    layers.Dense(10,  activation='softmax')         # Output layer: 10 classes + Softmax
], name='MNIST_Classifier')

# Step 6 — Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
# Step 7 — Train the Model
history = model.fit(
    X_train, y_train_ohe,
    epochs=10,
    batch_size=32,
    validation_split=0.1,
    verbose=1
)

# Evaluate on test set
test_loss, test_acc = model.evaluate(X_test, y_test_ohe, verbose=0)
print(f"\n{'='*40}")
print(f"  Test Accuracy : {test_acc*100:.2f}%")
print(f"  Test Loss     : {test_loss:.4f}")
print(f"{'='*40}")
```

```python
# Training curves
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(history.history['accuracy'],     label='Train', color='#4e91d4', lw=2)
axes[0].plot(history.history['val_accuracy'], label='Val',   color='#e05c5c', lw=2)
axes[0].set_title('Model Accuracy', fontweight='bold')
axes[0].set_xlabel('Epoch'); axes[0].set_ylabel('Accuracy')
axes[0].legend(); axes[0].grid(True, alpha=0.3)

axes[1].plot(history.history['loss'],     label='Train', color='#4e91d4', lw=2)
axes[1].plot(history.history['val_loss'], label='Val',   color='#e05c5c', lw=2)
axes[1].set_title('Model Loss', fontweight='bold')
axes[1].set_xlabel('Epoch'); axes[1].set_ylabel('Loss')
axes[1].legend(); axes[1].grid(True, alpha=0.3)

plt.suptitle('MNIST Neural Network — Training History', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('ex3_training_curves.png', dpi=150, bbox_inches='tight')
plt.show()
print("Plot saved ✓")
```

## Exercise 4 — Forward Propagation Calculation (House Prices)

```python
import numpy as np

# Given values
x1 = 2000   # Square Footage
x2 = 3      # Number of Bedrooms
w1 = 0.5    # Weight for Square Footage
w2 = 0.7    # Weight for Bedrooms
b  = 50000  # Bias

# Step 1 — Compute the weighted sum z (before activation)
z = (x1 * w1) + (x2 * w2) + b

# Step 2 — Apply ReLU activation: max(0, z)
def relu(z):
    return max(0, z)

prediction = relu(z)

print("=" * 50)
print("   FORWARD PROPAGATION — House Price Prediction")
print("=" * 50)
print(f"\nInputs:")
print(f"  x₁ (Square Footage) = {x1}")
print(f"  x₂ (Bedrooms)       = {x2}")
print(f"\nWeights & Bias:")
print(f"  w₁ = {w1},  w₂ = {w2},  b = {b:,}")
print(f"\nStep 1 — Weighted Sum (z):")
print(f"  z = (x₁×w₁) + (x₂×w₂) + b")
print(f"  z = ({x1}×{w1}) + ({x2}×{w2}) + {b:,}")
print(f"  z = {x1*w1:,.1f} + {x2*w2} + {b:,} = {z:,.1f}")
print(f"\nStep 2 — ReLU Activation:")
print(f"  ReLU({z:,.1f}) = max(0, {z:,.1f}) = {prediction:,.1f}")
print(f"\n{'='*50}")
print(f"  Predicted House Price: ${prediction:,.1f}")
print(f"{'='*50}")
print("""
INTERPRETATION:
  z = 1,000 + 2.1 + 50,000 = 51,002.1
  ReLU output = 51,002.1 (since z > 0, ReLU passes it through unchanged)

  The model predicts a house price of ~$51,002.
  Note: This is a very simplified single-neuron network.
  A real house price model would have many more features,
  layers, and would be trained on actual market data.
  The weights here are not trained — they're initial values.
""")
```

## Exercise 5 (Optional) — Forward and Backward Propagation in Python

```python
import numpy as np

# Initialize input data (features)
x = np.array([4, 80])   # 4 hours studied, previous test score: 80

# Initialize weights and bias
w = np.array([0.6, 0.3])  # Initial weights
b = 10                      # Initial bias

# Forward Propagation
def forward_propagation(x, w, b):
    z = np.dot(x, w) + b   # Weighted sum
    return z                 # Linear activation (regression task — no ReLU)

# Compute prediction
y_pred = forward_propagation(x, w, b)
y_true = 85  # Actual exam score

# Compute Loss (Mean Squared Error)
loss = 0.5 * (y_true - y_pred) ** 2

# Compute Gradients (Backward Propagation)
grad_w = -(y_true - y_pred) * x   # ∂Loss/∂w
grad_b = -(y_true - y_pred)        # ∂Loss/∂b

# Update Weights and Bias (Gradient Descent)
learning_rate = 0.01
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b

print("=" * 50)
print("    FORWARD + BACKWARD PROPAGATION")
print("    Predicting Exam Score from Study Hours")
print("=" * 50)
print(f"\nInputs:       x = {x}  (hours={x[0]}, prev_score={x[1]})")
print(f"Weights:      w = {w}")
print(f"Bias:         b = {b}")
print(f"\n--- FORWARD PASS ---")
print(f"z = dot({x}, {w}) + {b}")
print(f"z = {np.dot(x,w)} + {b} = {y_pred}")
print(f"\nPrediction:   ŷ = {y_pred:.2f}")
print(f"True score:   y = {y_true}")
print(f"Loss (MSE):   L = 0.5 × ({y_true} - {y_pred})² = {loss:.4f}")
print(f"\n--- BACKWARD PASS (Gradient Descent) ---")
print(f"∂L/∂w = {grad_w}")
print(f"∂L/∂b = {grad_b:.4f}")
print(f"\n--- WEIGHT UPDATE (lr = {learning_rate}) ---")
print(f"w_new = {w} - {learning_rate} × {grad_w} = {w_new}")
print(f"b_new = {b} - {learning_rate} × {grad_b:.4f} = {b_new:.4f}")

# Run multiple epochs to show learning
print(f"\n--- TRAINING FOR 20 EPOCHS ---")
w_cur, b_cur = np.array([0.6, 0.3]), 10.0
print(f"{'Epoch':<8}{'Prediction':<14}{'Loss':<12}{'w'}")
print("-" * 55)
for epoch in range(1, 21):
    pred  = forward_propagation(x, w_cur, b_cur)
    loss  = 0.5 * (y_true - pred) ** 2
    gw    = -(y_true - pred) * x
    gb    = -(y_true - pred)
    w_cur = w_cur - learning_rate * gw
    b_cur = b_cur - learning_rate * gb
    if epoch in [1, 5, 10, 15, 20]:
        print(f"  {epoch:<6}{pred:<14.4f}{loss:<12.4f}{np.round(w_cur, 4)}")

print(f"\nExplanation:")
print("""
  Each epoch, the gradient descent step pushes w and b in the
  direction that reduces the loss. The error (y_true - y_pred)
  shrinks every iteration because:
  - If prediction is too low  → gradients are negative → weights increase
  - If prediction is too high → gradients are positive → weights decrease
  
  The learning rate (0.01) controls how big each step is.
  Too large → overshooting; Too small → very slow convergence.
""")
```

## Exercise 6 (Optional) — Visualizing Predictions on MNIST

```python
# Steps 0–3 already done in Exercise 3 (model, X_test, y_test loaded)

# Step 7 — Make predictions on the test set
y_pred_proba = model.predict(X_test, verbose=0)
y_pred_class = np.argmax(y_pred_proba, axis=1)  # Predicted class index

# Step 8 — Visualize 20 predictions (correct = green, wrong = red)
fig, axes = plt.subplots(4, 5, figsize=(14, 11))
axes = axes.flatten()

indices = np.random.choice(len(X_test), 20, replace=False)

for i, idx in enumerate(indices):
    axes[i].imshow(X_test[idx], cmap='gray_r')
    true_label = y_test[idx]
    pred_label = y_pred_class[idx]
    confidence = y_pred_proba[idx][pred_label] * 100
    color = '#2ecc71' if pred_label == true_label else '#e05c5c'
    axes[i].set_title(
        f"True: {true_label}\nPred: {pred_label} ({confidence:.1f}%)",
        fontsize=9, color=color, fontweight='bold'
    )
    axes[i].axis('off')
    for spine in axes[i].spines.values():
        spine.set_edgecolor(color)
        spine.set_linewidth(3)

plt.suptitle('MNIST Predictions  |  Green = Correct  |  Red = Wrong',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('ex6_predictions.png', dpi=150, bbox_inches='tight')
plt.show()

# Summary stats
correct = np.sum(y_pred_class == y_test)
print(f"\nCorrect predictions: {correct}/{len(y_test)} ({correct/len(y_test)*100:.2f}%)")
print("Plot saved ✓")
```
