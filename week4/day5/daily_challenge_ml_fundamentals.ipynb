## 1. Qu'est-ce que le Machine Learning ?

Le **Machine Learning (ML)** est une branche de l'intelligence artificielle qui permet aux machines
d'**apprendre automatiquement a partir de donnees**, sans etre explicitement programmees pour chaque tache.

Plutot que d'ecrire des regles manuellement, on fournit des exemples au modele qui en extrait
des patterns pour faire des predictions ou prendre des decisions.

### Pourquoi le ML est essentiel pour les analystes de donnees ?

| Avantage | Description |
|----------|-------------|
| **Scalabilite** | Traite des millions de donnees en quelques secondes |
| **Generalisation** | Predit sur de nouvelles donnees jamais vues |
| **Automatisation** | Remplace des taches repetitives d'analyse manuelle |
| **Precision** | Detecte des patterns invisibles a l'oeil humain |
| **Adaptabilite** | Le modele s'ameliore avec plus de donnees |

L'analyste de donnees utilise le ML pour **transformer des donnees brutes en decisions actionnables**.

## 2. Applications du ML dans differents secteurs

### Secteur 1 : Sante / Medical
- **Detection de cancer** : un modele CNN analyse des images IRM et detecte des tumeurs
  avec une precision superieure aux radiologues dans certains cas.
- **Prediction de readmission** : les hopitaux predisent quels patients risquent d'etre
  readmis dans les 30 jours apres leur sortie pour mieux allouer les ressources.
- *Type ML utilise : Apprentissage supervise (classification/regression)*

### Secteur 2 : Finance / Banque
- **Detection de fraude** : les banques analysent chaque transaction en temps reel
  et bloquent automatiquement les transactions suspectes (ex: Visa, Mastercard).
- **Credit scoring** : algorithme qui evalue le risque de defaut de paiement d'un emprunteur
  a partir de son historique financier.
- *Type ML utilise : Apprentissage supervise + anomalie detection (non supervise)*

### Secteur 3 : E-commerce / Retail
- **Systemes de recommandation** : Amazon, Netflix, Spotify suggerent des produits/contenus
  personnalises bases sur le comportement de l'utilisateur.
- **Gestion des stocks** : prediction de la demande pour optimiser les niveaux de stock
  et reduire les ruptures ou surstock.
- *Type ML utilise : Apprentissage non supervise (clustering) + supervise (regression)*

### Secteur 4 : Transport / Automobile
- **Voitures autonomes** : Tesla, Waymo utilisent du ML (vision par ordinateur + RL)
  pour percevoir l'environnement et prendre des decisions de conduite en temps reel.
- **Optimisation des itineraires** : Google Maps predit le trafic et calcule le meilleur trajet.
- *Type ML utilise : Deep Learning + Apprentissage par renforcement*

### Secteur 5 : Agriculture
- **Detection de maladies des plantes** : analyse d'images de feuilles pour identifier
  les infections et alerter les agriculteurs avant propagation.
- **Irrigation intelligente** : prediction des besoins en eau selon la meteo et le type de sol.
- *Type ML utilise : Vision par ordinateur (CNN) + Regression*

### Resume des 3 types

| Type | Definition | Donnees | Exemple concret |
|------|-----------|---------|----------------|
| **Supervise** | Apprend a partir de donnees **etiquetees** (input → output connu) | Labelisees | Detecter si un email est spam |
| **Non Supervise** | Trouve des **structures cachees** dans des donnees sans labels | Non labelisees | Segmenter des clients par comportement |
| **Renforcement** | Un **agent** apprend par essais/erreurs en maximisant une **recompense** | Interactions | Robot qui apprend a marcher |

---
#### Apprentissage Supervise — Detail
- **Definition :** Le modele apprend une fonction `f(X) = y` a partir de paires (input, label).
- **Sous-types :** Classification (sortie categorielle) | Regression (sortie numerique)
- **Algorithmes :** Decision Tree, Random Forest, SVM, Regression Logistique, Neural Networks
- **Scenario :** Un hopital veut predire si un patient est diabetique (1/0) a partir de son
  age, IMC, glycemie. Les historiques de patients diagnostiques servent de donnees d'entrainement.

#### Apprentissage Non Supervise — Detail
- **Definition :** Le modele trouve des patterns sans labels, en regroupant ou reduisant les donnees.
- **Sous-types :** Clustering (K-Means, DBSCAN) | Reduction de dimension (PCA, t-SNE) | Detection d'anomalies
- **Algorithmes :** K-Means, Hierarchical Clustering, Autoencoders, Isolation Forest
- **Scenario :** Une banque groupe ses clients en segments (gros depensiers, economiseurs, etc.)
  sans savoir a l'avance combien de groupes existent, pour personnaliser les offres.

#### Apprentissage par Renforcement — Detail
- **Definition :** Un agent interagit avec un environnement, recoit des recompenses/penalites
  et apprend une politique optimale par essais et erreurs.
- **Composants :** Agent | Environnement | Etat | Action | Recompense | Politique
- **Algorithmes :** Q-Learning, Deep Q-Network (DQN), PPO, A3C
- **Scenario :** AlphaGo (DeepMind) a appris a jouer au jeu de Go en jouant des millions
  de parties contre lui-meme, atteignant un niveau surhumain.

## 4. Le Processus de Developpement d'un Modele ML

```python
# Organigramme du processus ML
fig, ax = plt.subplots(figsize=(14, 18))
fig.patch.set_facecolor('#0d1117')
ax.set_facecolor('#0d1117')
ax.set_xlim(0, 10)
ax.set_ylim(0, 18)
ax.axis('off')

def draw_box(ax, x, y, w, h, text, color, textcolor='white', fontsize=11, style='round,pad=0.1'):
    box = FancyBboxPatch((x - w/2, y - h/2), w, h,
                          boxstyle=style, facecolor=color,
                          edgecolor='white', linewidth=1.5)
    ax.add_patch(box)
    ax.text(x, y, text, ha='center', va='center', color=textcolor,
            fontsize=fontsize, fontweight='bold', wrap=True,
            multialignment='center')

def draw_arrow(ax, x1, y1, x2, y2, color='white'):
    ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                arrowprops=dict(arrowstyle='->', color=color, lw=2))

def draw_diamond(ax, x, y, w, h, text, color):
    diamond = plt.Polygon(
        [[x, y+h/2], [x+w/2, y], [x, y-h/2], [x-w/2, y]],
        closed=True, facecolor=color, edgecolor='white', lw=1.5
    )
    ax.add_patch(diamond)
    ax.text(x, y, text, ha='center', va='center', color='white',
            fontsize=9, fontweight='bold', multialignment='center')

# === ETAPES ===
steps = [
    (5, 17,   3.5, 0.8,  'PROBLEME METIER\nDefinir l objectif & les KPIs', '#e94560'),
    (5, 15.2, 3.5, 0.8,  'COLLECTE DES DONNEES\nSources : BDD, API, scraping, capteurs', '#0f3460'),
    (5, 13.4, 3.5, 0.8,  'NETTOYAGE & PREPARATION\nValeurs manquantes, outliers, doublons', '#16213e'),
]

highlight_steps = [
    (5, 11.2, 3.8, 1.0,  '1. SELECTION DES FEATURES\nCorrelation, importance, PCA, VIF', '#1a6b3c'),
    (5, 9.2,  3.8, 1.0,  '2. SELECTION DU MODELE\nBaseline, cross-validation, benchmark', '#1a3a6b'),
    (5, 7.2,  3.8, 1.0,  '3. EVALUATION DU MODELE\nMetriques : accuracy, F1, RMSE, AUC', '#6b1a1a'),
]

end_steps = [
    (5, 5.2,  3.5, 0.8,  'OPTIMISATION (TUNING)\nGridSearch, RandomSearch, Bayesian', '#4a4a00'),
    (5, 3.4,  3.5, 0.8,  'DEPLOIEMENT\nAPI, Docker, cloud (AWS/GCP/Azure)', '#2d4a2d'),
    (5, 1.6,  3.5, 0.8,  'MONITORING & MAINTENANCE\nDrift detection, re-entrainement', '#4a2d00'),
]

for (x, y, w, h, text, color) in steps + highlight_steps + end_steps:
    draw_box(ax, x, y, w, h, text, color)

# Fleches entre etapes
ys = [17, 15.2, 13.4, 11.2, 9.2, 7.2, 5.2, 3.4, 1.6]
for i in range(len(ys)-1):
    draw_arrow(ax, 5, ys[i]-0.4, 5, ys[i+1]+0.5)

# Fleche retour si modele insuffisant
ax.annotate('', xy=(5, 9.7), xytext=(5, 6.7),
            arrowprops=dict(arrowstyle='->', color='#ffe66d', lw=2,
                            connectionstyle='arc3,rad=-0.5'))
ax.text(7.8, 8.2, 'Modele\ninsuffisant\n→ re-selectionner', color='#ffe66d',
        fontsize=8, ha='center', style='italic')

# Labels des 3 etapes principales
for y, label, color in [(11.2, 'ETAPE 1', '#2ecc71'), (9.2, 'ETAPE 2', '#3498db'), (7.2, 'ETAPE 3', '#e74c3c')]:
    ax.text(1.0, y, label, color=color, fontsize=10, fontweight='bold', va='center')

ax.set_title('Processus Complet de Developpement d un Modele ML',
             color='white', fontsize=15, fontweight='bold', pad=15)

plt.tight_layout()
plt.savefig('plot2_processus_ml.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()
print('Organigramme sauvegarde.')
```

### Etape 1 : Selection des Features (Caracteristiques)

**Definition :** Identifier quelles variables d'entree sont les plus pertinentes pour predire la cible.

**Pourquoi c'est crucial ?**
- Trop de features → surapprentissage (overfitting), modele lent
- Features non pertinentes → bruit, mauvaise generalisation
- Bonne selection → modele plus simple, plus rapide, plus interpretable

**Methodes :**

| Methode | Description |
|---------|-------------|
| Correlation de Pearson | Mesure la relation lineaire entre feature et cible |
| Feature Importance | Score donne par les modeles d'arbres (Random Forest) |
| PCA | Reduit la dimension tout en preservant la variance |
| Variance Inflation Factor (VIF) | Detecte la multicollinearite entre features |
| SelectKBest | Selectionne les K meilleures features par test statistique |

**Exemple :** Pour predire le prix d'une maison, on garde : surface, localisation, nombre de chambres.
On elimine : identifiant proprietaire, code couleur facade (non informatifs).

---
### Etape 2 : Selection du Modele

**Definition :** Choisir l'algorithme le mieux adapte au probleme, aux donnees et aux contraintes.

**Criteres de choix :**

| Critere | Questions a se poser |
|---------|---------------------|
| Type de probleme | Classification ? Regression ? Clustering ? |
| Volume de donnees | Peu de data → modele simple. Beaucoup → deep learning |
| Interpretabilite | Besoin d'expliquer ? → Decision Tree, regression logistique |
| Performance | Vitesse vs precision ? |
| Dimensionalite | Beaucoup de features ? → SVM, neural networks |

**Processus :** Partir d'un **baseline** simple (regression lineaire, dummy classifier),
puis comparer avec des modeles plus complexes via **cross-validation (k-fold)**.

---
### Etape 3 : Evaluation du Modele

**Definition :** Mesurer objectivement la performance du modele sur des donnees **non vues** (test set).

**Metriques par type de probleme :**

| Probleme | Metriques principales |
|----------|----------------------|
| Classification | Accuracy, Precision, Rappel, F1-Score, AUC-ROC |
| Regression | MAE, MSE, RMSE, R² |
| Clustering | Silhouette Score, Davies-Bouldin Index |

**Pieges a eviter :**
- **Overfitting** : modele trop entraine sur le train set, echoue sur le test set
- **Data leakage** : informations du test set qui contaminent l'entrainement
- **Classe desequilibree** : accuracy trompeuse → privilegier F1 ou AUC

```python
# Visualisation : Comparaison de modeles et metriques
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.patch.set_facecolor('#0d1117')

# --- Graphique 1 : Bias-Variance Tradeoff ---
ax = axes[0]
ax.set_facecolor('#16213e')
complexity = np.linspace(1, 10, 100)
bias       = 5 / complexity
variance   = 0.5 * complexity**0.8
total      = bias + variance

ax.plot(complexity, bias,     '--', color='#ff6b6b', lw=2, label='Biais')
ax.plot(complexity, variance, '--', color='#4ecdc4', lw=2, label='Variance')
ax.plot(complexity, total,    '-',  color='#ffe66d', lw=2.5, label='Erreur totale')
idx_min = np.argmin(total)
ax.axvline(complexity[idx_min], color='white', ls=':', lw=1.5, label='Complexite optimale')
ax.set_facecolor('#16213e')
ax.set_title('Biais-Variance Tradeoff', color='white', fontweight='bold')
ax.set_xlabel('Complexite du modele', color='white')
ax.set_ylabel('Erreur', color='white')
ax.legend(fontsize=8, facecolor='#0d0d1a', labelcolor='white')
ax.tick_params(colors='white')
for spine in ax.spines.values(): spine.set_color('#333')

# --- Graphique 2 : Comparaison de modeles (Radar / Bar) ---
ax = axes[1]
ax.set_facecolor('#16213e')
modeles = ['Linear\nReg.', 'Decision\nTree', 'Random\nForest', 'SVM', 'Neural\nNet']
accuracy = [0.72, 0.78, 0.91, 0.85, 0.93]
bar_colors = ['#e74c3c', '#3498db', '#2ecc71', '#f39c12', '#9b59b6']
bars = ax.bar(modeles, accuracy, color=bar_colors, edgecolor='white', lw=0.8)
for bar, val in zip(bars, accuracy):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
            f'{val:.0%}', ha='center', va='bottom', color='white', fontsize=10, fontweight='bold')
ax.axhline(0.80, color='#ffe66d', ls='--', lw=1.5, label='Seuil acceptable (80%)')
ax.set_title('Comparaison de modeles (Accuracy)', color='white', fontweight='bold')
ax.set_ylabel('Accuracy', color='white')
ax.set_ylim(0.5, 1.05)
ax.legend(fontsize=8, facecolor='#0d0d1a', labelcolor='white')
ax.tick_params(colors='white')
for spine in ax.spines.values(): spine.set_color('#333')

# --- Graphique 3 : Matrice de confusion ---
ax = axes[2]
ax.set_facecolor('#16213e')
cm = np.array([[45, 5], [8, 42]])
im = ax.imshow(cm, cmap='Blues')
labels = [['VP\n45', 'FP\n5'], ['FN\n8', 'VN\n42']]
subtitles = [['(Vrais Positifs)', '(Faux Positifs)'], ['(Faux Negatifs)', '(Vrais Negatifs)']]
for i in range(2):
    for j in range(2):
        color = 'white' if cm[i,j] > 30 else 'black'
        ax.text(j, i-0.15, labels[i][j], ha='center', va='center',
                color=color, fontsize=14, fontweight='bold')
        ax.text(j, i+0.2, subtitles[i][j], ha='center', va='center',
                color=color, fontsize=7)
ax.set_xticks([0, 1])
ax.set_yticks([0, 1])
ax.set_xticklabels(['Predit: Positif', 'Predit: Negatif'], color='white')
ax.set_yticklabels(['Reel: Positif', 'Reel: Negatif'], color='white')
precision = 45/(45+5)
rappel    = 45/(45+8)
f1        = 2 * precision * rappel / (precision + rappel)
ax.set_title(f'Matrice de Confusion\nPrecision={precision:.0%} | Rappel={rappel:.0%} | F1={f1:.0%}',
             color='white', fontweight='bold')
for spine in ax.spines.values(): spine.set_color('#333')

plt.suptitle('Etape 3 : Evaluation du Modele', color='white', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('plot3_evaluation_modele.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()
print('Graphique 3 sauvegarde.')
```
